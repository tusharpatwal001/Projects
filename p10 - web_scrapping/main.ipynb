## 2. Read .csv from Multiple URLs with Pandas

```python
# link structure
leagues = ["0", "1", "2", "3", "C"]
frames = []
for league in leagues:
    df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2425/E"+ league +".csv")
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element

frames[:2]
```

```python
# all leagues with seasons
leagues = ["0", "1", "2", "3", "C"]
seasons = ["2021", "2122", "2223", "2324", "c425"]
frames = []
for season in seasons:
    for league in leagues:
        df = pd.read_csv("https://www.football-data.co.uk/mmz4281/"+ season +"/E"+ league +".csv", encoding="unicode_escape")
        frames.append(df)
```
